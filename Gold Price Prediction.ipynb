```python
##############################################   [Importing libraries]   ##############################################

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

#######################################  [Data Collection and Processing]   ############################################

gold_data=pd.read_csv('gld_price_data.csv') 
print("Data loaded successfully...")              # Loading the dataset
    


gold_data=gold_data.dropna()   
print("Cleaning Null Values if any...")                # Checking and cleaning null values

   


X=gold_data.drop(['Date','GLD'],axis=1)            #Seperating features and Target (date column is not required for prediction)
Y=gold_data['GLD']
print("Features and Target seperated successfully...")


###################################   [Splitting data into training and testing]   ####################################
 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)     # train contains 80% of data, test contains 20% of data 
print("Data seperated into training set and testing set...")                                 # X contains features and Y contains corresponding gold price


###################################   [Model Training - Random Forest Regressor]   ####################################

regressor=RandomForestRegressor(n_estimators=100)  
regressor.fit(X_train,Y_train)                         # Using 100 decision trees (RFG is a combination of multiple decision trees)
if(regressor):
    print("Random Forest Regressor model created and data trained successfully...")
                
#######################################  [Model Evalutaion]   ################################################

test_data_prediction=regressor.predict(X_test)                                     
print("predicted the output Successfully...")             # Predicting the gold price using the test datas features


#######################################  [Error Detection]   ################################################

print("Calculating the error in the model...")
error_score=metrics.r2_score(Y_test,test_data_prediction)  # R squared error
print("R squared error : ",error_score)

Y_test=list(Y_test)                                       # Converting Y_test to list
plt.plot(Y_test,color='red',label='Actual Value')        # Plotting the actual values
plt.plot(test_data_prediction,color='yellow',label='Predicted Value') # Plotting the predicted values
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()                                                 # Displaying the plot






```
